# Box-Cox transform

```python
import sys
sys.path.append('D:/source/repos')
from utilities.std_imports import *
import scipy.stats as ss
from pandas_datareader import data
```

```python
ts = data.get_data_yahoo('AAPL', '2017-01-01', '2018-01-01')['High']
plt.plot(ts);
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
prob = ss.probplot(ts, dist=ss.norm, plot=ax1)
ax1.set_xlabel('')
ax1.set_title('Probplot against normal distribution');
```

```python
ts_bcx, lmbda = ss.boxcox(ts)
print('Lambda = ', lmbda)
plt.plot(ts_bcx);
```

```python
fig = plt.figure()
ax2 = fig.add_subplot(111)
prob = ss.probplot(ts_bcx, dist=ss.norm, plot=ax2)
ax2.set_title('Probplot after Box-Cox transformation')
plt.show()
```
