# Liapunov exponent 
For predictability: the more positive, the more chaotic

```python
import sys
sys.path.append('D:/source/repos')
from utilities.std_imports import *
import nolds
from pandas_datareader import data
```

```python
ts = np.array(range(1,1000))
ts = ts * ts/2
plt.figure(figsize=[22,6])
plt.plot(ts);
le = max(nolds.lyap_e(ts))
print('Liapunov exponent = ', le)
```

```python
ts = data.get_data_yahoo('AAPL', '2017-01-01', '2018-01-01')['High']
plt.figure(figsize=[22,6])
plt.plot(ts);
le = max(nolds.lyap_e(ts))
print('Liapunov exponent = ', le)
```

```python
lm = nolds.logistic_map(0.1, 1000, r=4)
ts = np.fromiter(lm, dtype="float32")
plt.figure(figsize=[22,6])
plt.plot(ts);
le = max(nolds.lyap_e(ts))
print('Liapunov exponent = ', le)
```
