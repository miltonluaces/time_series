# Random walk

```python
from utilities.std_imports import *
from time_series.forecastability.random_walk import *
from mpl_toolkits.mplot3d import Axes3D 
import pylab 
import random 
```

### 1D

```python
rw = random_walk_1D(nSteps=500) 
plt.figure(figsize=[12,6])
plt.plot(rw[0], rw[1],'blue', label = "Randwalk1D") 
plt.title("1-D Random Walks")
```

### 2D

```python
x, y = random_walk_2D(nSteps=500)
plt.figure(figsize=[12,6])
pylab.title("Random Walk 2-D")
pylab.plot(x, y);
```

### 3D

```python
x, y, z = random_walk_3D(nSteps=500)
plt.figure(figsize=[20,10])
ax = plt.subplot(1,1,1, projection='3d')
ax.plot(x, y, z,alpha=0.9) #alpha sets the darkness of the path.
ax.scatter(x[-1],y[-1],z[-1]);
```
