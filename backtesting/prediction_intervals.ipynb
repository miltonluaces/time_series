Prediction interval: Given a prediction of ‘y’ given ‘x’, there is a 95% likelihood that the range ‘a’ to ‘b’ *covers* the true outcome.

### Prediction Interval Coverage Probability (PICP)
Percentage of true values that fell inside the interval

```python
ts = [12, 13, 15, 18, 22, 21, 15, 12, 8,  12, 15, 18, 22, 25, 27, 23, 21, 23, 21, 25, 18, 27, 25, 22]
fc = [13, 11, 15, 13, 29, 11, 14, 18, 12, 11, 15, 13, 15, 29, 23, 21, 29, 22, 21, 13, 11, 22, 23, 15]

plt.figure(figsize=[20,5])
plt.plot(ts)
plt.plot(fc);
```

```python
def get_alpha(ts, fcst, amp):
    hits = 0
    for t in range(len(ts)):
        lwr = ts[t] - amp
        upr = ts[t] + amp
        if fc[t]>=lwr and fc[t]<=upr: hits=hits+1
    alpha = hits/len(ts)
    return alpha
```

```python
for i in range(1, 13):
    alpha = get_alpha(ts=ts, fcst=fc, amp=i)
    print(round(alpha*100,2), '% :', '[f -',i, ', f +', i, ']')
```
