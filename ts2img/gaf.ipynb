# Gramian Angular Field

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
from pyts.image import GramianAngularField
from pyts.datasets import load_gunpoint
plt.rcParams['mpl_toolkits.legacy_colorbar'] = False
plt.rcParams['figure.figsize'] = (12,6)
```

Given a time series, build the G matrix calculating the dot products:

$\large X(t) = x_1, x_2, ..., x_t $

$\large
G = \;
\begin{bmatrix}
x_1 \cdot x_1 & x_1 \cdot x_2 & ... & x_1 \cdot x_t\\
x_1 \cdot x_2 & x_2 \cdot x_2 & ... & x_2 \cdot x_t\\
... & ... & ... & ...\\
x_t \cdot x_t & x_2 \cdot x_2 & ... & x_t \cdot x_t\\
\end{bmatrix}
$

### Load data

```python
X, _, _, _ = load_gunpoint(return_X_y=True)
X.shape
```

```python
plt.plot(X[12]);
```

### GAF transformation : summation

```python
gaf = GramianAngularField(image_size=24, method='summation')
X_gafSum = gaf.fit_transform(X)
X_gafSum.shape
```

```python
img_idx = 12
plt.imshow(X_gafSum[img_idx], cmap='rainbow', origin='lower')
plt.colorbar();
```

### GAF transformation : difference

```python
gaf = GramianAngularField(image_size=24, method='difference')
X_gafDiff = gaf.fit_transform(X)
X_gafDiff.shape
```

```python
img_idx = 12
plt.imshow(X_gafDiff[img_idx], cmap='rainbow', origin='lower')
plt.colorbar();
```
