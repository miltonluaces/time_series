# TS Autocorrelation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from random import randrange
from pandas import Series
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
ts = pd.read_csv('D:/data/csv/min_temp.csv', header=0, index_col=0)
plt.figure(figsize=(15,8))
plt.plot(ts)
plt.show()
```

```python
plot_acf(ts)
plt.show()
```

```python
plot_pacf(ts)
plt.show()
```

```python
pd.plotting.autocorrelation_plot(ts);
```

```python
df = pd.read_csv('D:/data/csv/milkProd.csv')
ts = df['MilkProd']
df.plot();
```

```python
def plot_autocorr(ts):
    fig, ax = plt.subplots(3, figsize=(16,10))
    ax[0].plot(ts)
    ax[1] = plot_acf(ts, ax=ax[1], lags=25)
    ax[2] = plot_pacf(ts, ax=ax[2], lags=25)
```

```python
plot_autocorr(ts)
```
