# Angle based outlier detection

```python
from pyod.models.abod import ABOD
from pyod.utils.data import generate_data
from pyod.utils.data import evaluate_print
from pyod.utils.example import visualize
```

### Generate data

```python
X_train, y_train, X_test, y_test = generate_data(n_train=200, n_test=100, n_features=2)
print('\nX')
print(X_train[:20])
print('\ny')
print(y_train)
```

### Training

```python
model = ABOD(method='fast')
model.fit(X_train)
```

### Validation

```python
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)
scores = model.decision_function(X_test)
proba = model.predict_proba(X_test)
```

```python
evaluate_print('ABOD', y_test, scores)
```

```python
visualize(clf_name=model, X_train=X_train, y_train=y_train, X_test=X_test, y_test=y_test, y_train_pred=pred_train, y_test_pred=pred_test, show_figure=True, save_figure=False)
```
