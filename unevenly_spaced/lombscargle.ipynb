# Lomb Scargle periodogram for unevenly spaced time series

```python
import sys
sys.path.append('D:/source/repos')
from utilities.std_imports import *
import scipy.signal as signal
```

### Create unevenly spaced time series

```python
A = 2.
w = 1.
phi = 0.5 * np.pi
nin = 1000
nout = 100000
sel_points = 0.9
```

```python
r = np.random.rand(nin)
t = np.linspace(0.01, 10 * np.pi, nin)
t = t[r >= sel_points]
```

```python
ts = A * np.sin(w * t + phi)
plt.figure(figsize=[12,6])
plt.plot(t, ts, '--', color='red');
plt.plot(t, ts, 'o');
```

### Generate periodogram

```python
frec = np.linspace(0.01, 10, nout)
pgram = signal.lombscargle(t, ts, frec, normalize=True)
plt.figure(figsize=[12,6])
plt.plot(frec, pgram);
```
