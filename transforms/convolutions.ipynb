# Convolutions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
dice = [1,2,3,4,5,6]

conv2 = np.convolve(dice, dice)
print(conv2)
plt.plot(conv2);
plt.show()

conv3 = np.convolve(conv2, dice)
print(conv3)
plt.plot(conv3);
plt.show()
```

### Moving average

```python
ts = [2,2,3,4,5,6,7,8,7,6,4,5,6,6,6,7,7,8,8,12,14,16,12,11,10,10,11,12,12]
mw = 3
W = np.repeat(1.0, mw)/mw
smas = np.convolve(ts, W, 'valid')
plt.plot(ts);
plt.plot(smas);
```

```python
sig = np.repeat([0., 1., 0.], 100)
win = signal.windows.hann(50)
filtered = signal.convolve(sig, win, mode='same') / sum(win)


fig, (ax_orig, ax_win, ax_filt) = plt.subplots(3, 1, sharex=True)
ax_orig.plot(sig)
ax_orig.set_title('Original pulse')
ax_orig.margins(0, 0.1)
ax_win.plot(win)
ax_win.set_title('Filter impulse response')
ax_win.margins(0, 0.1)
ax_filt.plot(filtered)
ax_filt.set_title('Filtered signal')
ax_filt.margins(0, 0.1)
fig.tight_layout()
fig.show()
```
