# TS Decomposition

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from random import randrange
from pandas import Series
```

### Random data

Additive descomposition

```python
ts = [i+randrange(10) for i in range(1,100)]
res = seasonal_decompose(ts, model='additive', period=1);
res.plot()
plt.show()
```

Multiplicative descomposition

```python
ts = [i**2.0 for i in range(1,100)]
res = seasonal_decompose(ts, model='multiplicative', period=1)
res.plot()
plt.show()
```
