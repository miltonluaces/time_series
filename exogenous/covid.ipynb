```python
from utilities.std_imports import *
plt.rcParams["figure.figsize"] = (26,10)
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
dates = pd.date_range(start=df.columns[4],end=df.columns[df.shape[1]-1])
df.head()
```

```python
df_de = df[df['Country/Region']=='Germany']
ts = pd.Series(df_de.iloc[:, 4:].values.flatten(), index=dates)
plt.plot(ts);
```

```python
diff = ts.diff(periods=1)
diff_roll = diff.rolling(90).mean()
plt.plot(diff_roll);
```
