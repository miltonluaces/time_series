# Categorical time series encoding

```python
import sys
sys.path.append('D:/source/repos')
from utilities.std_imports import *
import sklearn.preprocessing as skp
import seaborn as sb
```

### Load data

```python
values = ['A', 'B', 'A', 'A', 'C', 'B', 'A', 'A', 'A', 'B', 'B', 'C', 'A', 'C', 'A', 'C', 'C']
ts = pd.DataFrame({'time': list(range(len(ts))), 'value':values})
ts.index = ts.time
sb.catplot(x='time', y='value', data=ts);
```

### One hot encoding

```python
ts_s = np.array(ts).reshape(-1,1)
ohe = skp.OneHotEncoder()
ohe.fit(ts_s)
ts_oh = ohe.transform(ts_s).toarray().astype(int)
print(ts_oh)
```

### Multivariate

```python
ts_m = pd.DataFrame(ts_oh).T
ts_m
```

```python
for i in range(3):
    plt.plot(ts_m.loc[i])
    plt.show()
```
